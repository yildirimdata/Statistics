Before we take up the discussion of correlation, we need to examine a way **to display the relation between two variables x and y. The most common and easiest way is a scatter plot**.

A scatter plot of two variables shows the values of one variable on the Y axis and the values of the other variable on the X axis. SP reveals patterns in **bivariate data.**

**Independent variable is also called explanatory variable and it can be displayed on x axis. On the other hand, dependent variable is called response varable and showed on why axis. Response variable (dep var) measures the outcome of a study; and explanatory variable (indep var) explains the outcome of the study.**

**A scatter plot shows the direction of a relationship between the variables.** A clear direction happens when there is either:

* High values of one variable occurring with high values of the other variable or low values of one variable occurring with low values of the other variable.

* High values of one variable occurring with low values of the other variable.

You can determine the strength of the relationship by looking at the scatter plot and seeing how close the points are to a line, a power function, an exponential function, or to some other type of function. For a linear relationship there is an exception. Consider a scatter plot where all the points fall on a horizontal line providing a "perfect fit." The horizontal line would in fact show no relationship.

# Correlation and Covariance

**Covariance** shows the direction of the linear relationship between two variables. (Varyans tek bir deikenin deikenlii; kovaryans ise 2 veya daha fazala deikenin birden fazla ortak deiimi). 2 variables can covary in the same or opposite directions.

**Correlation** gives information about the direction and the strength of the linear relationship between two variables. 

**Correlation** is one of the most common statistical concepts. It is a statistical technique that determines how one variable changes with another variable. It **gives us the degree of the linear relationship between the two variables.**

**Meriam-Webster definition**: "A relation existing between phenomena or things or between mathematical or statistical variables which tend to vary, be associated, or occur together in a way not expected on the basis of chance alone."

For example, consider the correlation between height and weight variables. If taller people also have more weight it can be said there is a correlation between height and weight.



**Covariance** provides similar information with correlation. However, **correlation goes beyond covariance and gives information also about the strength of the relationship between two quantitative variables.** Covariance does not provide information about the strength of the relationship.

**Both can be positive or negative.** Correlation or covariance is positive if one increases the other also increases and negative if one increases the other decreases. The figure below shows the positive and negative correlations between the two variables.

# Pearson Correlation Coefficient

There are different methods to calculate the correlation coefficient between two variables. The most famous one is the Pearson Correlation Coefficient. It is a number between -1 and 1 that indicates the strength of the relationship.

-1 : Indicates complete negative correlation

+1 : Indicates complete correlation

0 : Indicates no correlation

# **Calculate Correlation and Covariance with Python**

# numpy.corrcoef and numpy.cov

We can easily calculate covariance and correlation with **numpy**. 

**The results of np.cov() and np.corrcoef()** commands are displayed in **matrix form**. 

Let's use the abovementioned covariance between temperature and number of people which was 24, and the correlation which was 0,9.

```python
#with numpy library
import numpy as np

temp=[93,84,82,78,98,70]

number_of_people=[13,10, 11, 8, 15, 9]

print("covariance: ", np.cov(temp, number_of_people))

print("correlation: ", np.corrcoef(temp, number_of_people))
```

```python
print("covariance is ", np.cov(temp, number_of_people)[0][0])
print("correlation is ", np.corrcoef(temp, number_of_people)[0][1])
```

# Pandas correlation calculation

**We can also use pandas series.corr() method to calculate the correlation coefficient**

syntax = serie_1.corr(serie2) 

It uses the pearsman correlation calculation as default. If we want to change the method: serie_1.corr(serie2, method = "spearman") or serie_1.corr(serie2, method = "kendall") or serie_1.corr(serie2, method = "any_callable_function"). Below an example.   

Complementary reading: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.corr.html

```python
import pandas as pd

x = pd.Series(range(10, 20))
y = pd.Series([2, 1, 4, 5, 8, 12, 18, 25, 96, 48])
```

```python
r = x.corr(y)
print("Pearson's correlation coefficient between x and y arrays is", r)
```

```python
# let's change it to spearman
rho = x.corr(y, method = "spearman")
print("Spearman's correlation coefficient between x and y arrays is", rho)
```

```python
# let's have an example with a callable method

def histogram_intersection(a, b):
    v = np.minimum(a, b).sum().round(decimals=1)
    return v

s1 = pd.Series([.2, .0, .6, .2])
s2 = pd.Series([.3, .6, .0, .1])
s1.corr(s2, method=histogram_intersection)
```

```python
a = pd.Series(range(21,39,3))
b = pd.Series([1,2,3,4,5])

# Important: pandas serie.corr(serie) function can calculate the correlation even if the length of two arrays
# don't match. In this case length a is 6 and length b is 5, however pandas' corr() returned the corr coef.
print(len(a))
print(len(b))
r = a.corr(b)
print(r)
```
