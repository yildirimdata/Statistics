# Sampling Distribution of the Sample Mean

Normally, the sampling distribution of the mean does not exist. It is theoretical. Because it is the distribution of the means we would get if we took infinite numbers of samples (every possible sample). That means we should use infinite numbers of samples.

In the diagram below, we can see that the population distribution is far from a normal distribution. A sample taken from the population leads to the sample mean x-bar. "All sample means" form the "sampling distribution of sample means". And the sampling distribution of the mean has a normal distribution form and it is narrower than the population distribution. This is explained by the Central Limit Theorem.



# Central Limit Theorem

The Central Limit Theorem states that as the sample size gets larger the sampling distribution of the sample means approaches a normal distribution — no matter the shape of the population distribution. This fact holds especially true if the sample sizes (n) over 30. Because, statisticians have found that for many population distributions, a sample size of at least 30 is “large enough.”

3 kuralı var:

- µ = µxbar (pop mean = sample mean)
- std = std / np.sqrt(n) yani s
- sample size (n) > 30 ise SDofSM normal distribution olur.

Merkeze limitlendigi icin central limit theory denmis. Tum hipotez testlerinin arkasinda bu vardir.



# Sampling Error and Confidence Intervals

# Sampling Error

Taking only one sample and obtaining the results of a sample statistic, such as the mean, creates a point estimate of the population parameter. If a different sample is selected, this single estimate will almost certainly not be the same. Because the sample statistics differ from sample to sample.

Remember the 10 million babies and 50 samples with a sample size of 100 (n=100). Sample means (the average weight of 100 babies) of each sample will differ from sample to sample. For example, in the first group, it might be 3.5 kg and in the second group 3.6 kg, etc. Imagine it varies from 3.2 to 3.8 kg. Some of the sample means will be higher than the population mean and some of the sample means will be lower than the population mean. **In statistics, this expected variation of the sample means is explained by sampling error.**

In that case, we can say that the **sampling error is: "the error that arises as a result of taking a sample from a population rather than using the whole population".**

The amount of the sampling error is primarily based on the variation in the population itself and on the size of the sample selected (n). **Larger samples (when n is greater) will have less sampling error but will be more costly to study on.**

**We can't avoid sampling error but we can estimate and reduce the error. One way is margin of error. If we enlarge the size of sample, margin error will reduce. But this can be very costy. Another way is good sample design. For example creating a sample which contains representatives of all subpopulations in the main population. These can reduce but can't avoid sampling error. In order to avoid it completely, we have to assess each observations in a population, but this inpractical and impossible. Therefore, we should take into accoun that the samples and surveys are not perfect but an efficient way to get inferences for a population.**

# Standard Error

Standard deviation of the sampling distribution of the mean

The Standard Error (SE) is similar to the standard deviation. They are both measures of spread. The two terms are actually equal with one important difference. **While the standard deviations use population data, the standard error uses sample data.**

In statistics, we very often come across the term “the standard error of the mean”. **The standard error tells us how far our sample statistic (like the sample mean) deviates from the actual population mean.** The larger your sample size (n) leads to the smaller the standard error.

# Confidence Interval

In practice, for estimating a population parameter (for example population mean) only one sample is used as the basis. But, taking into consideration the differences among several samples (if we repeat an experiment or survey over and over again), statisticians have developed the concept of a **"confidence interval"** which indicates **the likelihood that a stated interval with a lower and upper limit properly estimates the parameter**.



The confidence interval concept is always used together with the confidence level. The confidence level is expressed as a percentage (for example, a 90%, 95% or 99% confidence level). It means that you should repeat an experiment or survey over and over again, 90 percent of the time your results will match the results you get from a population. And the confidence interval is the gap between lower and upper limits.

**Confidence Interval Formula**



xbar = sample mean
z = z-score (calculated from confidence level)
𝜎 = standard deviation 
n = sample size 

For example, you survey a group of pet owners to see how many cans of cat food they buy a year. The sample mean is 125, sample size is 25, sample standard deviation is 30. Just imagine you calculated this interval at the 90% confidence level and get a confidence interval of 125±10 = [115, 135]. We can say that with 90% confidence, they buy between 115 and 135 cans a year in average. That means we can be 90% confident that the population mean is somewhere between 115 and 135, based on only 25 samples.

We can calculate narrower interval (with same confidence level) as sample size increases. In our previous example, if we increase the sample size from 25 to 100, the 90% confidence interval will be approximately between [120, 130].

Two things which affect confident intervals:

1) The less variation in a population leads to a narrower confident intervalt, vice versa. Bec population with low variation leads to samples with low variation.

2) Bigger sample size lower CIs.

# t Distribution

The t distribution (or Student’s t Distribution) is a family of distributions that look very similar to the normal distribution curve, only a bit shorter and fatter. **The t distribution is used to estimate population parameters when the sample size is small and/or when the population standard deviation (σ) is unknown.** 

**The larger the sample size, the more the t distribution approaches the normal distribution. For sample sizes larger than 30, the distribution is very similar to the normal distribution.**

- t is more spread out than z. It is "flatter" than normal. This is because the use of s introduces more uncertainty, making t "sloppier" than z.
- The amount of spread depends on the sample size. The greater the sample size, the more confident we can be that s is near σ, and the closer t gets to z, the normal.
- Degrees of freedom (dof or d.f.): Given n pieces of data, you use up one "degree of freedom" when you compute x̄, leaving n-1 independent pieces of information.

**As the degrees of freedom (n-1) increases, the t-distribution will get closer to the standard normal distribution (z-distribution).**

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
xs = np.linspace(-5,5,1000)

plt.plot(xs, stats.t.pdf(xs, 1), "b", label="t (df=1)")
plt.plot(xs, stats.t.pdf(xs, 3), "g", label="t (df=3)")
plt.plot(xs, stats.t.pdf(xs, 8), "c", label="t (df=8)")
plt.plot(xs, stats.t.pdf(xs, 15), "m", label="t (df=15)")
plt.plot(xs, stats.norm.pdf(xs), "k", label="Normal")
plt.legend()
plt.show();
```

# Confidence Interval for Population Mean using SciPy

# Large Sample or Population Variance Known (using z-score)

When working with large samples (n≥30), we can assume that the sampling distribution of the sample mean is approximately normally distributed (thanks to the Central Limit Theorem). In this case, we can calculate the confidence interval for the mean using the **stats.norm.interval()** function in Scipy.

Define a sample dataset (exam grades for a class of 40 students):

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
grades = np.random.randint(100, size = 40)
print(grades)
```

```python
# Calculate the sample mean:
xbar = np.mean(grades)
print(xbar)
```

```python
# Calculate standart error of the mean:
n = len(grades)            #sample size
sigma = 28                 #population standard deviation (from past experiences)
sem = sigma / np.sqrt(n)   #standard error of the mean
print(sem)
```

```python
stats.sem(grades)
```

```python
# Calculate 95% Confidence Interval around the mean by using scipy.stats:

stats.norm.interval(confidence=0.95, loc=xbar, scale=sem)
```

# Population Variance Unknown (using t-score)

**Where the population standard deviation σ is unknown, the sample standard deviation s is used. This introduces extra uncertainty, as s varies from sample to sample. So, the t distribution is used instead of the normal distribution.** In this case, we can calculate the confidence interval for the mean using the **stats.t.interval()** function in Scipy.

```python
np.random.seed(101)
grades = np.random.randint(100, size = 20)
print(grades)
```

```python
# Calculate 95% Confidence Interval around the mean by using scipy.stats:

stats.t.interval(confidence=0.95,              #confidence level
                 df=np.size(grades)-1,    #degrees of freedom
                 loc=np.mean(grades),     #sample mean
                 scale=stats.sem(grades)  #standard error of the mean
)
```

**Population std (ya da variance) biliniyorsa ya da SS > 30 ise z test: stats.norm.interval(confidence, loc = xbar, scale= sem) (sem = sigma / np.sqrt(n) veya stats.sem())**

**Eğer pop std (ya da variance) bilinmiyor ve sample size kücük ise t scroe kullanırız. stats.t.interval(confidence, df= np.size()-1, loc = np.mean(), scale = stats.sem)**
